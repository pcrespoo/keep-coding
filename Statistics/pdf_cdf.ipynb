### Probability Density Function(PDF)
- This curve corresponds to the percentage of the distribution that belongs to a given bin(it ranges from 0 to 0.5)

### Cumulative Density Function
- This curve derives from PDF and corresponds to the cumulative sum of percentages of the distribution (it ranges from 0 to 1)

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('train.csv')
```

```python
fig,axes = plt.subplots(1,2,figsize=(10,8))
a = sns.distplot(df.Age.dropna(),ax=axes[0])
a.set_title('Histogram and PDF')
b = sns.distplot(df.Age.dropna(),hist_kws={'cumulative':True},kde_kws={'cumulative':True},ax=axes[1])
b.set_title('CDF')
```
